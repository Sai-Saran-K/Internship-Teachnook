```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=50, n_features=2, n_informative = 2, n_redundant = 0, n_clusters_per_class = 1, flip_y = 0, class_sep = 3, random_state=20)
```

```python
y[y == 0] = -1
```

```python
X.shape
```

```python
plt.scatter(X[:,0][y == 1], X[:,1][y == 1], color = 'blue')
plt.scatter(X[:,0][y == -1], X[:,1][y == -1], color = 'red')
plt.show
```

```python
def sign(x):
    return 2*(x>=0)-1
```

```python
iters = 10
w = np.zeros(X.shape[1])
for iter in range(iters):
    mistakes = 0
    for i in range(X.shape[0]):
        if sign(X[i].T @ w)!=y[i]:
            mistakes+=1
            w = w + y[i] * X[i]
    if mistakes == 0:
        print(f'Iteration: {iter}, Mistakes: {mistakes}')
        break
    print(f'Iteration: {iter}, Mistakes: {mistakes}')
```

```python
plt.scatter(X[:,0][y == 1], X[:,1][y == 1], color = 'blue')
plt.scatter(X[:,0][y == -1], X[:,1][y == -1], color = 'red')
x_db = np.linspace(-2,2)
y_db = -w[0]/w[1]*x_db
plt.plot(x_db, y_db)
plt.ylim(-4,-1)
plt.show()
```
